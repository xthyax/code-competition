```python
import cv2
import os
import glob
import numpy as np
import matplotlib.pyplot as plt
import random
import json
from tqdm import tqdm
from function_test import resize_image
from augmentation_setup import custom_augment, complex_augmentation, test_augmentation
```

```python
IMG_PATH = "D:\\Coding_practice\\_Data\\SIIM-ISIC_data\\train"

imgs_list = []
for image in glob.glob(IMG_PATH +  "\\*jpg"):
    imgs_list.append(image)

print(len(imgs_list))
```

```python
PATH = "D:\\Coding_practice\\_Data\\SIIM-ISIC_data"

OUTPUT_FOLDER = "train_compressed"

OUTPUT_PATH = os.path.join(PATH, OUTPUT_FOLDER)

os.makedirs(OUTPUT_PATH, exist_ok=True)
```

```python
columns = 2
rows = 5
fig = plt.figure(figsize=(20,20))
for i in range(1, columns * rows + 1):
    img = cv2.imread(random.choice(imgs_list))
    img, _ ,_,_,_= resize_image(img,min_dim=256, max_dim=512)
    # print(img.shape)
    # aug_img = complex_augmentation(image=img)
    fig.add_subplot(rows, columns, i)
    plt.imshow(img)
plt.show()
```
